```python
%%writefile app.py
#Creamos el archivo de la APP en el interprete principal (Phyton)

#Verificamos que todas las librerias se puedan importar
#Importamos librerias
import streamlit as st
import base64
import plotly.express as px
import pandas as pd
import numpy as np
from funpymodeling.exploratory import freq_tbl
import seaborn as sns
import matplotlib.pyplot as plt
import math
import folium
from folium.plugins import HeatMap
from collections import Counter
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression


#Definimos la instancia de streamlit
@st.cache_resource

#################################################################################################

# Funciónes para mostrar la imagen en la sidebar usando HTML
def get_base64_of_bin_file(file_path):
    with open(file_path, 'rb') as binary_file:
        binary_content = binary_file.read()
        return base64.b64encode(binary_content).decode('utf-8')

def display_image_in_sidebar(file_path):
    base64_image = get_base64_of_bin_file(file_path)
    img_html = f'<img src="data:image/png;base64,{base64_image}" width="1000%" />'
    st.sidebar.markdown(img_html, unsafe_allow_html=True)

#Creamos la función de carga de datos
def load_data():
    #Lectura del archivo csv sin indice
    df1=pd.read_csv("RIO DE JANEIRO sin outliers.csv")
    df2=pd.read_csv("RIO DE JANEIRO sin outliers.csv", index_col= 'review_scores_rating')
    df2=df2.sort_index(ascending=False)
    #Carga desde un archivo csv con indice
    df3=pd.read_csv("RIO DE JANEIRO dashboard.csv")

#################################################################################################

    #Etapa de procesamiento de Datos

    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table= freq_tbl(df1['host_is_superhost'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro= table[table['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index1= Filtro.set_index('host_is_superhost')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index1
    numeric_df1= Filtro_index1.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols1= numeric_df1.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table2= freq_tbl(df1['room_type'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro2= table2[table2['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index2= Filtro2.set_index('room_type')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index2
    numeric_df2 = Filtro_index2.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols2= numeric_df2.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table3= freq_tbl(df1['property_type'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro3= table3[table3['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index3= Filtro3.set_index('property_type')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index3
    numeric_df3 = Filtro_index3.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols3= numeric_df3.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table4= freq_tbl(df1['host_verifications'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro4= table4[table4['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index4= Filtro4.set_index('host_verifications')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index4
    numeric_df4 = Filtro_index4.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols4= numeric_df4.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table5= freq_tbl(df1['amenities'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro5= table5[table5['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index5= Filtro5.set_index('amenities')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index5
    numeric_df5 = Filtro_index5.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols5= numeric_df5.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table6= freq_tbl(df1['instant_bookable'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro6= table6[table6['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index6= Filtro6.set_index('instant_bookable')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index6
    numeric_df6 = Filtro_index6.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols6= numeric_df6.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table7= freq_tbl(df1['host_neighbourhood'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro7= table7[table7['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index7= Filtro7.set_index('host_neighbourhood')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index7
    numeric_df7 = Filtro_index7.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols7= numeric_df7.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table8= freq_tbl(df1['has_availability'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro8= table8[table8['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index8= Filtro8.set_index('has_availability')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index8
    numeric_df8 = Filtro_index8.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols8= numeric_df8.columns                          #Devuelve lista de Columnas



    #ANÁLISIS UNIVARIADO DE FRECUENCIAS
    #Obtengo un análisis univariado de una variable categórica en específico
    table9= freq_tbl(df1['host_location'])
    #Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
    Filtro9= table9[table9['frequency']>1]
    #Ajusto el indice de mi dataframe
    Filtro_index9= Filtro9.set_index('host_location')

    #Selecciono las columnas tipo numericas del dataframe Filtro_index9
    numeric_df9 = Filtro_index9.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols9= numeric_df9.columns



    #Selecciono las columnas tipo numericas del dataframe df2
    numeric_df10 = df2.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols10 = numeric_df10 .columns                 #Devuelve lista de Columnas

    #Selecciono las columnas tipo numericas del dataframe df3
    numeric_df11 = df3.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols11= numeric_df11.columns                 #Devuelve lista de Columnas

    return Filtro_index1, Filtro_index2, Filtro_index3, Filtro_index4, Filtro_index5, Filtro_index6, Filtro_index7, Filtro_index8, Filtro_index9, df1, df2, df3, numeric_df1, numeric_df2, numeric_df3, numeric_df4, numeric_df5, numeric_df6, numeric_df7, numeric_df8, numeric_df9, numeric_df10, numeric_df11, numeric_cols1, numeric_cols2, numeric_cols3, numeric_cols4, numeric_cols5, numeric_cols6, numeric_cols7, numeric_cols8, numeric_cols9, numeric_cols10, numeric_cols11

#Cargo los datos obtenidos de la función "load_data"
Filtro_index1, Filtro_index2, Filtro_index3, Filtro_index4, Filtro_index5, Filtro_index6, Filtro_index7, Filtro_index8, Filtro_index9, df1, df2, df3, numeric_df1, numeric_df2, numeric_df3, numeric_df4, numeric_df5, numeric_df6, numeric_df7, numeric_df8, numeric_df9, numeric_df10, numeric_df11, numeric_cols1, numeric_cols2, numeric_cols3, numeric_cols4, numeric_cols5, numeric_cols6, numeric_cols7, numeric_cols8, numeric_cols9, numeric_cols10, numeric_cols11 = load_data()

######################CREACIÓN DEL DASHBOARD################################

#1. CREACIÓN DE LA SIDEBAR

# Configuración de la página
st.set_page_config(
    page_title="Dashboard Río de Janeiro",
    page_icon="📊",
    layout="wide",
    initial_sidebar_state="expanded",
)

# Bloque de CSS para el tema oscuro y la barra lateral
css = """
<style>
/* Fondo oscuro para la barra lateral */
section[data-testid="stSidebar"] {
    background-color: #001f3f;
}

/* Cambiar el color del texto en la barra lateral */
section[data-testid="stSidebar"] * {
    color: #F5F5DC !important;
}

/* Estilo para checkboxes, radio buttons, select boxes y multiselect */
section[data-testid="stSidebar"] .st-bx, /* checkboxes */
section[data-testid="stSidebar"] .st-bu, /* radio buttons */
section[data-testid="stSidebar"] .st-bo, /* select boxes */
section[data-testid="stSidebar"] .stMultiSelect { /* multiselect */
    background-color: #white !important; /* Verde esmeralda */
    color: #001f3f !important; /* Texto oscuro */
    padding: 8px; /* Aumenta el padding para más espacio */
    border-radius: 5px;
    height: auto; /* Ajusta la altura automáticamente */
    min-height: 36px; /* Establece una altura mínima */
}

/* Estilo de las listas desplegables */
section[data-testid="stSidebar"] .stSelect,
section[data-testid="stSidebar"] .stMultiSelect {
    background-color: #white !important; /* Fondo verde */
    color: #001f3f !important; /* Texto oscuro */
}

/* Estilo del botón */
section[data-testid="stSidebar"] .stButton > button {
    background-color: #50C878 !important; /* Verde esmeralda */
    color: #001f3f !important; /* Texto oscuro */
    border: none; /* Sin borde */
    border-radius: 5px; /* Bordes redondeados */
    padding: 10px 20px; /* Espaciado interno */
    cursor: pointer; /* Cursor de mano */
}

/* Estilo del botón en hover */
section[data-testid="stSidebar"] .stButton > button:hover {
    background-color: #45a049 !important; /* Verde más oscuro en hover */
}

/* Mejorar la visibilidad del texto dentro de entradas de texto */
section[data-testid="stSidebar"] .st-dx {
    background-color: #white !important;
    color: #F5F5DC !important;
}
</style>
"""

# Insertar el CSS personalizado en la aplicación
st.markdown(css, unsafe_allow_html=True)

# Ruta de la imagen que acabas de subir
file_path = "IM1.png"  # Asegúrate de que el nombre coincida con el archivo subido

# Mostrar la imagen en la sidebar
display_image_in_sidebar(file_path)

# Encabezados y texto en la sidebar
st.sidebar.markdown(
    """
    <div style="background-color: #001f3f; padding: 10px;">
        <h2 style="color: #F5F5DC;">DASHBOARD RIO DE JANEIRO</h2>
        <h3 style="color: #F5F5DC;">Panel de Control</h3>
    </div>
    """,
    unsafe_allow_html=True
)

# CREACIÓN DE LOS FRAMES
# Generamos los Frames que utilizaremos en el diseño
#Widget 1: Selectbox
#Menu desplegable de opciones de los Frames seleccionados
Frames= st.selectbox(label= "Opciones de Navegación", options= ["Distribución Geográfica de Airbnb", "Análisis Univariado", "Análisis de Correlación", "Análisis de Regresión"])

# Definir tamaño de la figura
fig_size = (10, 6)


# 1. CONTENIDO DEL FRAME 1
# Generamos los encabezados para el dashboard
if Frames == "Distribución Geográfica de Airbnb":
    st.title("RIO DE JANEIRO")
    st.header("Mapa de calor")

    # Botón en la sidebar para ver el mapa
    show_map = st.sidebar.button("Ver Mapa")

    if show_map:
        # Mapeo de coordenadas
        coordinates = {
            'Copacabana': [-22.9711, -43.1822],
            'Barra da Tijuca': [-22.9942, -43.3666],
            'Ipanema': [-22.9834, -43.2046],
            'Recreio dos Bandeirantes': [-22.9911, -43.2650],
            'Leblon': [-22.9804, -43.1749],
            'Botafogo': [-22.9216, -43.1933],
            'Centro': [-22.9070, -43.1860],
            'Jacarepaguá': [-22.9007, -43.2445],
            'Santa Teresa': [-22.9172, -43.1732],
            'Flamengo': [-22.9261, -43.1741],
            'Leme': [-22.7421, -43.1315],
            'Vidigal': [-22.9745, -43.1851],
            'Laranjeiras': [-22.9150, -43.2250],
            'Tijuca': [-22.9270, -43.3191]
        }

        # Contar frecuencias para host_neighbourhood y host_location
        neighbourhood_frequencies = Counter(df1['host_neighbourhood'])
        location_frequencies = Counter(df1['host_location'])

        # Crear DataFrames para las tablas
        neighbourhood_df = pd.DataFrame(neighbourhood_frequencies.items(), columns=['Neighbourhood', 'Frequency']).sort_values(by='Frequency', ascending=False)
        location_df = pd.DataFrame(location_frequencies.items(), columns=['Location', 'Frequency']).sort_values(by='Frequency', ascending=False)

        # Crear columnas para mostrar las tablas lado a lado
        col1, col2 = st.columns(2)

        with col1:
            st.write("*Tabla de Frecuencias de Vecindarios:*")
            st.dataframe(neighbourhood_df)

        with col2:
            st.write("*Tabla de Frecuencias de Ubicaciones:*")
            st.dataframe(location_df)

        # Crear el mapa base
        m = folium.Map(location=[-22.9068, -43.1729], zoom_start=12)

        # Añadir datos de calor
        heat_data = []
        for neighbourhood, freq in neighbourhood_frequencies.items():
            if neighbourhood in coordinates:
                lat, lon = coordinates[neighbourhood]
                heat_data.append([lat, lon, freq])  # Añadir frecuencia

        # Crear mapa de calor con frecuencia
        HeatMap(heat_data, radius=15).add_to(m)

        # Mostrar el mapa en Streamlit
        st.components.v1.html(m._repr_html_(), height=500)


# 2. CONTENIDO DEL FRAME 2
if Frames == "Análisis Univariado":
    st.title("RIO DE JANEIRO")
    st.header("Gráficas Univariadas")

    # Lista de los filtros ya creados
    Filtros_indices = {
        'host_is_superhost': Filtro_index1,
        'room_type': Filtro_index2,
        'property_type': Filtro_index3,
        'host_verifications': Filtro_index4,
        'amenities': Filtro_index5,
        'instant_bookable': Filtro_index6,
        'has_availability': Filtro_index8,
    }

    # Checkbox para mostrar el dataset
    check_box = st.sidebar.checkbox(label="Mostrar Dataset", value=False)

    # Checkbox para mostrar estadísticas descriptivas de df1
    stats_box = st.sidebar.checkbox(label="Mostrar Estadísticas Descriptivas", value=False)

    # Mostrar el dataset si el checkbox está activado
    if check_box:
        st.markdown('<p style="color: black;">Aquí está el Dataset:</p>', unsafe_allow_html=True)
        st.table(df2)

    # Mostrar estadísticas descriptivas si el checkbox está activado
    if stats_box:
        st.write("**Estadísticas Descriptivas de df1:**")
        st.table(df1.describe())

    # Multiselect para seleccionar categorías
    selected_categories = st.sidebar.multiselect(
        'Selecciona una o más variables categóricas:',
        options=list(Filtros_indices.keys())
    )

    # Si no se selecciona ninguna categoría, mostrar un mensaje
    if not selected_categories:
        st.sidebar.warning("Por favor, selecciona al menos una variable para analizar.")
    else:
        # Iterar sobre las categorías seleccionadas
        for category in selected_categories:
            Filtro_index = Filtros_indices[category]

            # Mostrar tabla de frecuencias
            st.subheader(f'Frecuencias de {category}')

            # Crear columnas para la tabla y la gráfica
            col1, col2 = st.columns([1, 2])  # Proporción de columnas

            with col1:
                st.table(Filtro_index)

            with col2:
                # Filtrar para mostrar solo los 15 más frecuentes
                if len(Filtro_index) > 15:
                    Filtro_index = Filtro_index.nlargest(15, 'frequency')

                # Graficar según la categoría seleccionada
                if category in ['room_type', 'host_is_superhost', 'instant_bookable', 'has_availability']:
                    colors = plt.cm.Paired(range(len(Filtro_index)))
                    plt.figure(figsize=(10, 5))
                    wedges, texts, autotexts = plt.pie(
                        Filtro_index['frequency'],
                        labels=None,
                        autopct=lambda pct: ('%1.1f%%' % pct) if pct > 1 else '',
                        pctdistance=0.85,
                        startangle=90,
                        colors=colors,
                        explode=[0.1] * len(Filtro_index)
                    )

                    plt.legend(wedges, Filtro_index.index, title="Categorías", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
                    plt.title(f'Distribución de {category}', fontsize=16, fontweight='bold')

                    for text in autotexts:
                        text.set_color('black')
                        text.set_fontsize(8)
                        text.set_weight('bold')

                    for wedge in wedges:
                        wedge.set_edgecolor('white')

                    plt.gca().set_facecolor('lightgray')
                    st.pyplot(plt)

                else:
                    colors = plt.cm.Paired(range(len(Filtro_index)))
                    plt.figure(figsize=(10, 5))
                    bars = plt.bar(Filtro_index.index, Filtro_index['frequency'], color=colors, edgecolor='black')

                    for bar in bars:
                        height = bar.get_height()
                        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom', fontsize=9, color='black')

                    plt.title(f'Distribución de {category}', fontsize=16, fontweight='bold')
                    plt.xlabel('Categorías', fontsize=14)
                    plt.ylabel('Frecuencia', fontsize=14)
                    plt.xticks(rotation=45, ha='right')
                    plt.grid(axis='y', linestyle='--', alpha=0.7)

                    st.pyplot(plt)

# 4. CONTENIDO DEL FRAME 3
# Generamos los encabezados para el dashboard
if Frames == "Análisis de Correlación":
    st.title("RIO DE JANEIRO")
    st.subheader("Scatter Plot, Heatmap y Análisis de Correlación")

    # Lista de variables numéricas
    numeric_vars = [
        'host_response_rate', 'host_acceptance_rate', 'accommodates',
        'price', 'number_of_reviews', 'review_scores_rating',
        'calculated_host_listings_count', 'reviews_per_month',
        'host_total_listings_count'
    ]

    # Selectbox para seleccionar la primera variable
    var1 = st.sidebar.selectbox('Selecciona la primera variable numérica:', options=numeric_vars)

    # Selectbox para seleccionar la segunda variable
    var2 = st.sidebar.selectbox('Selecciona la segunda variable numérica:', options=numeric_vars)

    # Asegurarse de que las variables seleccionadas son diferentes
    if var1 == var2:
        st.sidebar.warning("Por favor, selecciona dos variables diferentes para analizar.")

    check_box = st.sidebar.checkbox(label="Mostrar Dataset")
    if check_box:
        st.markdown('<p style="color: black;">Aquí está el Dataset:</p>', unsafe_allow_html=True)
        st.table(df2)

    else:

        # Crear gráfico de dispersión
        plt.figure(figsize=(10, 5))
        plt.scatter(df1[var1], df1[var2], alpha=0.6, edgecolors='w', color='blue', label='Datos')

        # Añadir línea de regresión
        sns.regplot(x=df1[var1], y=df1[var2], scatter=False, color='red', ci=None, label='Línea de Regresión')

        # Títulos y etiquetas
        plt.title(f'Gráfico de Dispersión: {var1} vs {var2}', fontsize=16, fontweight='bold')
        plt.xlabel(var1, fontsize=14)
        plt.ylabel(var2, fontsize=14)
        plt.grid(alpha=0.5)
        plt.legend()  # Añadir leyenda

        # Mostrar la gráfica
        st.pyplot(plt)

        # Análisis de correlación
        correlation = df1[[var1, var2]].corr().iloc[0, 1]
        st.subheader("Análisis de Correlación")
        st.write(f"La correlación entre {var1} y {var2} es de *{correlation:.2f}*.",
                 unsafe_allow_html=True)

        # Visualización de la matriz de correlación
        st.subheader("Heatmap de Correlación de Todas las Variables")
        correlation_matrix = df1[numeric_vars].corr()
        plt.figure(figsize=(10, 6))
        sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
        plt.title("Matriz de Correlación de Variables Numéricas", fontsize=16, fontweight='bold')
        st.pyplot(plt)



if Frames == "Análisis de Regresión":
    st.title("RIO DE JANEIRO")
    st.subheader("Análisis de Regresión con Diferentes Modelos")

    numeric_vars2 = [
        'host_response_rate', 'host_acceptance_rate', 'accommodates',
        'price', 'number_of_reviews', 'review_scores_rating',
        'calculated_host_listings_count', 'reviews_per_month',
        'host_total_listings_count'
    ]

    variable_1_4 = st.sidebar.selectbox("Selecciona la Variable 1 (Independiente)", options=numeric_vars2)
    variable_2_4 = st.sidebar.selectbox("Selecciona la Variable 2 (Dependiente)", options=numeric_vars2)

    if variable_1_4 == variable_2_4:
        st.sidebar.warning("Por favor, selecciona dos variables diferentes para analizar.")

    functions = ['Cuadrática', 'Exponencial', 'Logarítmica', 'Senoidal', 'Valor Absoluto']
    selected_functions = st.sidebar.multiselect("Selecciona las funciones a graficar:", functions)

    results = {func: None for func in functions}  # Inicializar el diccionario

    if variable_1_4 and variable_2_4:
        valid_indices = df1[[variable_1_4, variable_2_4]].dropna().index
        x = df1.loc[valid_indices, variable_1_4].values
        y = df1.loc[valid_indices, variable_2_4].values

        # Reemplazar NaN e Inf con un valor adecuado (por ejemplo, la media)
        x = np.nan_to_num(x, nan=np.nanmean(x), posinf=np.nanmean(x), neginf=np.nanmean(x))
        y = np.nan_to_num(y, nan=np.nanmean(y), posinf=np.nanmean(y), neginf=np.nanmean(y))

        plt.figure(figsize=(10, 6))
        plt.scatter(x, y, color='blue', label='Datos')

        # Crear máscara para valores válidos
        mask = ~np.isnan(x) & ~np.isinf(x) & ~np.isnan(y) & ~np.isinf(y)

        for func in functions:  # Calcular todos los R^2
            try:
                if func == 'Cuadrática':
                    X = np.column_stack((x[mask], x[mask]**2))
                    model = sm.OLS(y[mask], sm.add_constant(X)).fit()
                    results[func] = model.rsquared
                    if func in selected_functions:
                        plt.plot(x[mask], model.predict(sm.add_constant(X)), label='Ajuste Cuadrático', color='red')

                elif func == 'Exponencial':
                    valid_exp_indices = y > 0
                    if np.any(valid_exp_indices):
                        X = x[valid_exp_indices]
                        Y = np.log(y[valid_exp_indices])
                        model = sm.OLS(Y, sm.add_constant(X)).fit()
                        results[func] = model.rsquared
                        if func in selected_functions:
                            plt.plot(x[valid_exp_indices], np.exp(model.predict(sm.add_constant(X))), label='Ajuste Exponencial', color='orange')

                elif func == 'Logarítmica':
                    valid_log_indices = x > 0
                    if np.any(valid_log_indices):
                        X = np.log(x[valid_log_indices])
                        y_valid = y[mask][valid_log_indices]
                        model = sm.OLS(y_valid, sm.add_constant(X)).fit()
                        results[func] = model.rsquared
                        if func in selected_functions:
                            plt.plot(x[valid_log_indices], model.predict(sm.add_constant(X)), label='Ajuste Logarítmico', color='green')

                elif func == 'Senoidal':
                    X = np.column_stack((np.sin(x[mask]), np.cos(x[mask])))
                    model = sm.OLS(y[mask], sm.add_constant(X)).fit()
                    results[func] = model.rsquared
                    if func in selected_functions:
                        plt.plot(x[mask], model.predict(sm.add_constant(X)), label='Ajuste Senoidal', color='purple')

                elif func == 'Valor Absoluto':
                    X = np.abs(x[mask])
                    model = sm.OLS(y[mask], sm.add_constant(X)).fit()
                    results[func] = model.rsquared
                    if func in selected_functions:
                        plt.plot(x[mask], model.predict(sm.add_constant(X)), label='Ajuste Valor Absoluto', color='magenta')

            except Exception as e:
                results[func] = None
                st.warning(f'Error al ajustar el modelo para {func}: {e}')

        plt.title(f'Ajustes de Funciones para {variable_2_4} vs {variable_1_4}')
        plt.xlabel(variable_1_4)
        plt.ylabel(variable_2_4)
        plt.legend()
        st.pyplot(plt)

        # Crear un DataFrame con todos los resultados
        results_df = pd.DataFrame(list(results.items()), columns=['Función', 'R Cuadrado'])
        st.write("Resultados del Análisis de Regresión:")
        st.table(results_df)
```
